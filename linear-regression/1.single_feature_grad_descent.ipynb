```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Step 1: Prepare data
data_df = pd.read_csv("./data/ex1data1.txt", header=None)
data_arry = data_df.values
training_set_size = np.size(data_arry,1) #size of the training set

features_n_output = np.split(data_arry, 2, 1)
raw_features = features_n_output[0] #raw features
ones_arry = np.ones((np.size(raw_features, 0), 1)) #to account for theta0
features = np.concatenate((ones_arry, raw_features), axis=1)
expected_op = features_n_output[1] #expected output
```

```python
# Step 2: Visualize data
plt.scatter(raw_features, expected_op)
```

```python
# Step 3: Define hypothesis and cost function

#initialize model params to all zeros
mdl_params = np.zeros((np.size(features,1),1))

#predict the output using the params without any training
hypothesis = np.dot(features, mdl_params) 

#visualize the training data and model
plt.scatter(raw_features, expected_op);
plt.scatter(raw_features, hypothesis);
```

```python
#Step 4: Train model
def compute_cost(mdl_params, features, expected_op): 
    size = np.size(features, 0) 
    cost = 1/(2*size) * np.square(np.dot(features, mdl_params) - expected_op).sum() 
    return cost

def train_gradient_descent(mdl_params, features, expected_op, learning_rate, max_iterations):
    for count in range(1, max_iterations) : 
        size = np.size(features, 0) 
        feat_count = np.size(features, 1) 
        hypothesis = np.dot(features, mdl_params)
        mdl_params0 = mdl_params[0,0] - (1/size) * learning_rate *  np.sum((hypothesis - expected_op) * features[:,0].reshape(np.size(features,0), 1))
        mdl_params1 = mdl_params[1,0] - (1/size) * learning_rate *  np.sum((hypothesis - expected_op) * features[:,1].reshape(np.size(features,0), 1))
        mdl_params = np.array([mdl_params0, mdl_params1]).reshape(2, 1)
        cost = compute_cost(mdl_params, features, expected_op) 
        print("iteration #{}. Cost is {}. Params={}".format(count, cost, mdl_params)) 
    return mdl_params

#Train model using gradient descent 
max_iterations=1500 #define max iterations 
learning_rate = 0.01 #define learning rate 
print("initial cost is {}".format(compute_cost(mdl_params, features, expected_op)))
trained_mdl_params = train_gradient_descent(mdl_params, features, expected_op, learning_rate, max_iterations)

```

```python
#visualize the training data and model 
plt.scatter(raw_features, expected_op) 
hypothesis = np.dot(features, trained_mdl_params)  #recompute the hypothesis
hypothesis_line = plt.plot(raw_features, hypothesis)
plt.setp(hypothesis_line, color='r', linewidth=2.0)
```

```python
#Step 5: Predict 

def predict(input):
    #update input vectors with ones for the first model param
    ones_arry = np.ones((np.size(input, 0), 1)) #to account for theta0
    mod_input = np.concatenate((ones_arry, input), axis=1)
    output = np.dot(mod_input, trained_mdl_params)    
    return output

print("For x=7.4764, expected 4.3483 and got {}".format(predict([[7.4764]])))
print("For x=5.5277, expected 9.1302 and got {}".format(predict([[5.5277]])))
print("For x=8.5172, expected 4.2415 and got {}".format(predict([[8.5172]])))
print("For x=13.394, expected 9.0551 and got {}".format(predict([[13.394]])))
```
